## Get catalog data

```python
base_url = "https://opendata.cbs.nl/ODataCatalog//"
```

```python
r = requests.get("?".join((base_url, "$format=json"))).json()['value']
```

```python
r
```

```python
catalog_urls = {
    item["name"]: "?".join((base_url+item["url"], "$format=json"))
    for item in r
}
```

```python
catalog_urls
```

```python
tables = pd.DataFrame(requests.get(catalog_urls['Tables']).json()['value']).set_index("ID")
themes = pd.DataFrame(requests.get(catalog_urls['Themes']).json()['value']).set_index("ID")
tables_themes = pd.DataFrame(requests.get(catalog_urls['Tables_Themes']).json()['value']).set_index("ID")

# tables = requests.get(catalog_urls['Tables']).json()['value']
# themes = requests.get(catalog_urls['Themes']).json()['value']
# tables_themes = requests.get(catalog_urls['Tables_Themes']).json()['value']
```

## Arrange for KUMU

### Elements

```python
elements_tables = tables[['Identifier', 'ShortTitle']]
```

```python
elements_tables.columns = ["Label", "Description"]
```

```python
elements_tables.head()
```

```python
elements_themes = pd.DataFrame(themes['Title']).rename(columns={"Title": "Label"})
```

```python
elements_themes['Description'] = "A CBS Theme"
elements_themes
```

```python
elements = pd.concat((elements_tables, elements_themes)).reset_index(drop=True)
elements
```

### Relations

```python
tables_themes.head()
```

```python
tables_themes = tables_themes.merge(themes[["Title"]], left_on="ThemeID", right_index=True)
tables_themes = tables_themes.rename(columns={"Title": "ThemeTitle"})
tables_themes
```

```python
relations_tables = tables_themes[['TableIdentifier', 'ThemeTitle']]
relations_tables = relations_tables.rename(columns={'TableIdentifier': 'From', 'ThemeTitle': 'To'})
relations_tables['Description'] = "Table/Theme"
relations_tables
```

```python
themes.head()
```

```python
child_themes = themes[~themes["ParentID"].isnull()]
child_themes['ParentID'] = child_themes['ParentID'].astype("int32")
child_themes
```

```python
child_themes = child_themes.merge(themes[['Title']], left_on='ParentID', right_index=True).rename(columns={"Title_x": "Title", "Title_y": "ParentTitle"})
child_themes
```

```python
relations_themes = child_themes[['Title', 'ParentTitle']]
relations_themes = relations_themes.rename(columns={"Title": "From","ParentTitle": "To"})
relations_themes['Description'] = "Theme/Parent Theme"
relations_themes
```

```python
relations = pd.concat((relations_tables, relations_themes)).reset_index(drop=True)
relations
```

## Export as json

```python
import json
```

```python
elements_dict = elements.to_dict('records')
relations_dict = relations.to_dict('records')
```

```python
with open("kumu_data/v3_catalog.json", "w+") as f:
    json.dump({"elements": elements_dict, "connections": relations_dict}, f)
```
